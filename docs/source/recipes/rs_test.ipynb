```python
import pybmds
import pandas as pd
%matplotlib inline
from pybmds.datasets.transforms.rao_scott import RaoScott, Species

dataset = pybmds.DichotomousDataset(
    doses=[0, 15, 50, 150],
    ns=[149, 146, 142, 140],
    incidences=[0, 0, 2, 6],
)

rs = RaoScott(
    dataset = dataset,
    species = Species.rabbit,
)

display(rs.summary_df())
fig = rs.figure()
```

```python
report = rs.to_docx()
report.save("Rao_Scott_test_report.docx")
```
