```python
import pybmds
import pandas as pd
%matplotlib inline
from pybmds.datasets.transforms.rao_scott import RaoScott, Species

# Example where one or more dose group(s) has zero incidence

dataset = pybmds.DichotomousDataset(
    doses=[0, 15, 50, 150],
    ns=[149, 146, 142, 140],
    incidences=[0, 0, 2, 6],
)

rs = RaoScott(
    dataset = dataset,
    species = Species.rabbit,
)

display(rs.summary_df())
fig = rs.figure()
```

```python
report = rs.to_docx()
report.save("Rao_Scott_test_report_has_zero_incidence.docx")
```

```python
# Example where no dose group has zero incidence

dataset1 = pybmds.DichotomousDataset(
    doses=[0, 15, 50, 150],
    ns=[149, 146, 142, 140],
    incidences=[1, 2, 2, 6],
)

rs1 = RaoScott(
    dataset = dataset1,
    species = Species.rabbit,
)

display(rs1.summary_df())
fig1 = rs1.figure()
```

```python
report = rs1.to_docx()
report.save("Rao_Scott_test_report_no_zero_incidence.docx")
```
